```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.exponential(scale=1.0, size=1000)
#This pulls 1000 random numbers from an exponential distribution
```

```python
plt.figure(figsize=(8, 5))
plt.hist(data, bins=100, color='lightcoral', edgecolor='black')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Exponentially Distributed Random Numbers")
plt.grid(True)

plt.savefig("exponential_histogram.pdf")
plt.show()
```
