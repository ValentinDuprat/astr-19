```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 100)
#This creates a Numpy array from 0 to 1 using 100 steps
```

```python
def sin_func(x):
    return np.sin(x)

def cos_func(x):
    return np.cos(x)
    #Defining functions that return sin(x) and cos(x)
```

```python
# Calculate values
y_sin = sin_func(x)
y_cos = cos_func(x)

# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

# Left panel: sin(x)
axs[0].plot(x, y_sin, 'b')
axs[0].set_title("sin(x)")
axs[0].set_xlabel("x")
axs[0].set_ylabel("sin(x)")

# Right panel: cos(x)
axs[1].plot(x, y_cos, 'r')
axs[1].set_title("cos(x)")
axs[1].set_xlabel("x")
axs[1].set_ylabel("cos(x)")

# Adjust layout and save
plt.tight_layout()
plt.savefig("sin_cos_plot.pdf")
plt.show()
```
